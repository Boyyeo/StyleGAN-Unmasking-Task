```python
unmasked_dir = "/content/sample_data/AAFD_test_unmasked/"
masked_dir = "/content/sample_data/AAFD_test_masked/"
count = 0
for unmasked_image_dir in os.listdir(unmasked_dir):
  print(count)
  create_dir = "/content/sample_data/AAFD_test_unmasked_folder/" + unmasked_image_dir.split(".")[0] + "/"
  masked_from_dir = "/content/sample_data/AAFD_test_masked/" + unmasked_image_dir.split(".")[0] + ".jpg"
  if not os.path.exists(masked_from_dir):
    continue
  count += 1
  masked_to_dir = "/content/sample_data/AAFD_test_masked_folder/"
  move_dir  = unmasked_dir+unmasked_image_dir
  #print(create_dir)
  #print(move_dir)
  os.mkdir(create_dir)
  shutil.move(unmasked_dir+unmasked_image_dir,create_dir)
  shutil.move(masked_from_dir,masked_to_dir)
  if(count>=5000):
    break
```
